# Lab | Customer Analysis Round 3. EDA

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_csv('marketing_customer_analysis.csv')
data
```

```python
data.columns = data.columns.str.lower()
data.columns = data.columns.str.replace(' ', '_')
```

Task 1 Show DataFrame info.

```python
data.info()
```

Task 2 Describe DataFrame.

```python
data.describe()
```

Task 3 Show a plot of the total number of responses.

```python
fig, ax = plt.subplots(1,2, figsize=(8,4))
ax[0].bar(x = data['response'].value_counts().index, height = data['response'].value_counts().values, color = ['blue', 'red'])
ax[0].set_xlabel("Response")
ax[0].set_ylabel("Count")
ax[0].set_title("Total Number of Responses")


sns.countplot(data=data, x="response", order=['No','Yes'], palette=['blue', 'red'], ax = ax[1])
ax[1].set_xlabel("Response")
ax[1].set_ylabel("Count")
ax[1].set_title("Total Number of Responses")

plt.tight_layout()
plt.show()
```

Task 4 Show a plot of the response rate by the sales channel.

```python
fig, ax = plt.subplots()
sns.countplot(data=data, x="sales_channel",  palette=['blue', 'red'], hue="response")
plt.show()
```

Task 5 Show a plot of the response rate by the total claim amount.

```python
fig, ax = plt.subplots()
sns.barplot(data=data, x='response', y='total_claim_amount', palette=['blue', 'red'], ax = ax, hue='response')
plt.show()
```

Task 6 Show a plot of the response rate by income.

```python
fig, ax = plt.subplots()
sns.barplot(data=data, x='response', y='income', palette=['blue', 'red'], ax = ax)
plt.show()
```
